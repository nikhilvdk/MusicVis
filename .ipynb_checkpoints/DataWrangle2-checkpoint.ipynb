```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import sqlite3
```

```python
lyrics = pd.read_csv('metrolyrics.csv',index_col=0)
lyrics = lyrics[pd.notnull(lyrics['artist'])]
lyrics = lyrics[pd.notnull(lyrics['lyrics'])]
lyrics['artist'] = lyrics['artist'].str.replace('-', ' ')
lyrics['song'] = lyrics['song'].str.replace('-', ' ')
lyrics['lyrics'] = lyrics['lyrics'].str.replace('\n', ' ')
lyrics.head()
```

```python
%%capture
lyrics.drop_duplicates(subset='song', keep=False)
lyrics.drop_duplicates(subset='artist', keep=False)
```

```python
db = sqlite3.connect("database.sqlite")
reviews = pd.read_sql_query("select * from reviews,content where reviews.reviewid=content.reviewid;", db)
reviews.drop_duplicates(subset='artist', keep=False)
```

```python
test = lyrics.merge(reviews,on='artist')
test.head(200)
```

```python
free = test.drop_duplicates(subset=['artist'], keep='first')
```

```python
df = free[['song','year','artist','genre','lyrics','score','best_new_music','pub_year','content']]
```

```python
writer = pd.ExcelWriter('output.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
