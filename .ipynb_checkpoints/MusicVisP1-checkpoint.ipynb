```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('finaloutput.xlsx')
```

```python
from textblob import TextBlob
from textblob.np_extractors import FastNPExtractor
extractor = FastNPExtractor()
```

```python
newdf = pd.DataFrame()
```

```python
for i in df.index:
    reviewtxt = TextBlob(df.loc[i,'content'])
    df.loc[i,'review_polarity'] = reviewtxt.sentiment.polarity*100
    df.loc[i,'review_subj'] = reviewtxt.sentiment.subjectivity*100
    lyrics = TextBlob(df.loc[i,'lyrics'])
    df.loc[i,'lyrics_polarity'] = lyrics.sentiment.polarity*100
    df.loc[i,'lyrics_subj'] = lyrics.sentiment.subjectivity*100
    
```

```python
for i in df.index:
    curradjcnt =0
    currlyr = df.loc[i,'lyrics']
    currlyrblob = TextBlob(currlyr)
    for word, pos in currlyrblob.tags:
        if pos=="JJ" or pos=="JJS" or pos=="JJR":
            curradjcnt+=1
    df.loc[i,'lyrics_adj_count']=curradjcnt    
```

```python
for i in df.index:
    if df.loc[i,'score']<6 and df.loc[i,'score']>=5:
        totaltext8+=df.loc[i,'content']
```

```python
score8 = TextBlob(totaltext8)
```

```python
adjs = []
for word, pos in score8.tags:
    if pos=="JJ" or pos=="JJS" or pos=="JJR":
        adjs.append(word)
```

```python
adjtxt = ""
for word in adjs:
    adjtxt+= word+" "
```

```python
adjblob = TextBlob(adjtxt)
adjblobcount = adjblob.word_counts
from collections import Counter
c = Counter(adjblobcount)
c.most_common(10)
```

```python
howmany =0
lyricstext = ""
for i in df.index:
    if df.loc[i,'score']<2 and df.loc[i,'score']>=1:
        lyricstext+=df.loc[i,'song']
        howmany+=1
        
lyricsblob = TextBlob(lyricstext)
lyricsadj =""
for word, pos in lyricsblob.tags:
    if pos=="JJ" or pos=="JJS" or pos=="JJR":
        lyricsadj += word + " "
        
adjblob = TextBlob(lyricsadj)

adjcount = Counter(adjblob.word_counts)

adjcount.most_common(10)
```

```python
lyricscount = Counter(lyricsblob.word_counts)
print(sum(lyricscount.values()))
print(sum(adjcount.values()))
print(sum(adjcount.values())/sum(lyricscount.values()) * 100)
```
